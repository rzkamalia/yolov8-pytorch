## **Training Detection**

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8s.yaml').load('yolov8s.pt')  # build from YAML and transfer weights

# Train the model
model.train(data="/content/drive/MyDrive/Colab Notebooks/yolov8-training/aquarium/data.yaml", epochs=100)
```
